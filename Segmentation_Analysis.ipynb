```
#FriedMan Test
from scipy.stats import friedmanchisquare
import pandas as pd

df = pd.read_excel('/content/FreedMan.xlsx')

stat, p_value = friedmanchisquare(df['3DAttention_Rating'], df['3DMedSam_Rating'], df['ReconNet_Rating'], df['ResUNet_Rating'], df['VNet_Rating'])
print(f"Friedman Chi-Square Statistic: {stat}")
print(f"P-Value: {p_value}")
```
